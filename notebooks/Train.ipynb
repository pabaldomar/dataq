# **Import Modules:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# **Load Dataset:**

```python
raw_data = pd.read_csv('../data/train.csv')
df = raw_data.copy()
df
```

# **Check Dataset:**

```python
print(df.head())
```

```python
df.info()
```

# **Dataset Cleaning**

## Check Dataset for NULL value:

```python
df.isna().sum()    # df.isnull().sum()
```

## View NULL value:

```python
df[df['Postal Code'].isna()]     # can also used 'isnull()'
```

## Fill 'Postal Code' NULL:

```python
df['Postal Code'] = df['Postal Code'].fillna('05401')
df.info()
```

```python
df.isna().sum()
```

## Format "Order Date", "Ship Date" and "Postal Code" data type:

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Order Date']
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
df['Ship Date']
```

```python
df['Postal Code'] = df['Postal Code'].astype(int).astype(str)
df['Postal Code']
```

```python
df.info()
```

## Create "Order Year" column from "Order Date":

```python
df['Order Year'] = df['Order Date'].dt.year
df.head()
```

# **Dataset Grouping:**

## Yearly Sales Trend:

```python
yearly_sales_trend = df.groupby('Order Year', as_index=False).agg({'Sales': 'sum'})
yearly_sales_trend
```

### Yearly Sales Trend Visualization:

```python
#sns.set_style('whitegrid')
sns.lineplot(x='Order Year',
             y='Sales',
             data=yearly_sales_trend)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

## Product Sold the Most in 2018:

### Create DataFrame in year 2018:

```python
df_2018 = df[df['Order Year'] == 2018]
df_2018.head()
```

### Groupby 2018 Sales by "Product Name":

```python
prod_grp_2018 = df_2018.groupby('Product Name', as_index=False).agg({'Sales': 'sum'})
prod_grp_2018.head()
```

### Sort Product Sold in 2018:

```python
prod_grp_2018 = prod_grp_2018.sort_values(by='Sales', ascending=False)
prod_grp_2018.head()
```

#### Most Product Sold in 2018 Visualization:

```python
top_product = prod_grp_2018[prod_grp_2018['Sales'] >= 4416.174]
sns.set_style('whitegrid')
sns.barplot(y='Product Name',
            x='Sales',
            data=top_product,
            errorbar=('ci', False),
            orient='h',
            color='grey')
plt.title('Top Products in 2018')
plt.ylabel(' ')
plt.show()
```
