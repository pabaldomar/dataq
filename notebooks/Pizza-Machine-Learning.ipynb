# **Import Libraries**

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Load DataSet**

```python
raw_data = pd.read_csv('../data/Cleaned-Survey-Pizza.csv')
df = raw_data.copy()
df.head()
```

# **Machine Learning**

## **Data Preparation**

```python
X = df.drop(['Pizza_Rating'], axis=1)
y = df['Pizza_Rating']
```

## **Data Splitting: Train & Test**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

## **Building Model**

```python
model = DecisionTreeClassifier()
```

## **Training the Model**

```python
model.fit(X_train, y_train)
```

## **Applying Model to make Prediction**

```python
predictions = model.predict(X_test)
```

## **Evaluate Model Performance**

```python
score = accuracy_score(y_test, predictions)
score
```

## **Decision Tree Plotting**

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model,
               feature_names=X.columns,
               class_names=['0', '1'],
               label='all',
               filled=True,
               rounded=True)
plt.title('Pizza Machine Learning Decision Tree',
          fontdict={'fontsize': 20, 'fontweight': 'bold'})
# plt.savefig('/content/drive/My Drive/Colab Notebooks/Pizza-Machine-Learning-Decision-Tree.png', dpi=200)
plt.show()
```
